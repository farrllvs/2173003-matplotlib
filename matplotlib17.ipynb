```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Cek versi
print(matplotlib.__version__)
print(np.__version__)
```

**Predefined Styles**

```python
plt.style.available #melihat style apa saja yang tersedia
```

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

**Menerapkan multiple style**

```python
plt.style.use(['seaborn', 'dark_background'])
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

**Menerapkan temporary styling**

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')

plt.show()
```
