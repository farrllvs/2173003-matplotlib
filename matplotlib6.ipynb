```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

**Simple Pie Plot**

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
plt.pie(data, # digunakan untuk membuat pieplot
        labels=kategori, # optional
        autopct='%1.1f%%', # optional, menampilkan persentase
        startangle=90) # optional, untuk mengatur rotasi

plt.title('Contoh Pie Plot')

plt.show()
```

**Pengaturan Legend dan Color**

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue'] # Gunakan string
```

```python
plt.pie(data, 
        labels=kategori,
        colors=warna, # tambahkan parameter(str) ini untuk mengganti warna setiap bagian.
        autopct='%1.1f%%',
        startangle=90)

plt.legend() # menampilkan legend
plt.title('Contoh Pie Plot')

plt.show()
```

**Exploded Pie Plot**

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08] # nilai 0.0 berarti tidak ditarik keluar dari lingkaran
```

```python
plt.pie(data, 
        labels=kategori,
        colors=warna,
        explode=explode_var, # gunakan ini untuk mengatur jarak potongan
        autopct='%1.1f%%',
        startangle=90)

plt.title('Contoh Exploded Pie Plot')

plt.show()
```
