**Belajar Violin Plot | Violin Chart**

**import modules**

```python
%matplotlib inline
import matplotlib
```

```python
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

**Sample Dataset**

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200) #loc melambangkan mean, scale melambangkan std dev dan size melambangkan jumlah data point
data
```

**Simple Violin Plot**

```python
plt.violinplot(data) #function untuk membentuk violin plot

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
               showextrema=True, #extrema mengindikasikan garis tengah untuk min dan max
               showmeans=False, #means untuk menampilkan mean
               showmedians=False, # median untuk menampilkan nilai tengah
               quantiles=[0.25, 0.5, 0.75]) # q1, q2, q3

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

**Horizontal Violin Plot**

```python
plt.violinplot(data, vert=False) #membuat data menjadi horizontal

plt.title('Horizontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

**Multiple Violin Plot**

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
```

```python
plt.violinplot(data)

plt.title('Multiple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
