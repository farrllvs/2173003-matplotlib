```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

**Sample dataset**

```python
n = np.linspace(-3.0, 3.0, 100) # (Start, End, Amount)

x, y = np.meshgrid(n, n) #menggunakan meshgrid
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
(100, 100)
(100, 100)
(100, 100)
```

**3D Wireframe**

```python
fig = plt.figure()
ax = plt.axes(projection='3d') #menghasilkan 3d

ax.plot_wireframe(x, y, z, color='orange') #melakukan plotting wireframe

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

**3D Surface Plot**

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# ax.plot_surface(x, y, z, color='orange')
# ax.plot_surface(x, y, z, color='pink')
ax.plot_surface(x, y, z, cmap='viridis') #menggunakan plot surface untuk menghasilkan surface yang solid dan dapat memilih warna

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot')

plt.show()
```
